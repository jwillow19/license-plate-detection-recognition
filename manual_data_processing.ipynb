```python
blur_bar = pred_results['blur'].value_counts().plot(kind='bar', title='blur ratio')
blur_bar.set(xlabel="0 or 1", ylabel="value counts")

```

```python
occluded_bar = pred_results['occluded'].value_counts().plot(kind='bar', title='occluded ratio')
occluded_bar.set(xlabel="0 or 1", ylabel="value counts")
```

```python
nolp_bar = pred_results['no_lp'].value_counts().plot(kind='bar', title='No License Plate ratio')
nolp_bar.set(xlabel="0 or 1", ylabel="value counts")
```

```python
small_bar = pred_results['small'].value_counts().plot(kind='bar', title='Small Input ratio')
small_bar.set(xlabel="0 or 1", ylabel="value counts")
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12,10))

x = np.array([0,1])

axs[0, 0].bar(x, list(pred_results['blur'].value_counts()))
axs[0, 0].set_title('blur ratio')

axs[0, 1].bar(x, list(pred_results['occluded'].value_counts()))
axs[0, 1].set_title('occluded ratio')

axs[1, 0].bar(x, list(pred_results['no_lp'].value_counts()))
axs[1, 0].set_title('no LP ratio')

axs[1, 1].bar(x, list(pred_results['small'].value_counts()))
axs[1, 1].set_title('small input ratio')


for ax in axs.flat:
    ax.set(xlabel='0 or 1', ylabel='# observation')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
